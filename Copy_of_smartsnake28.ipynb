```python
"""
-----------------------------------------------------------------
SMART SNAKE AI DUEL ‚Äì Rule Based Agents (Short & Simple Version)
-----------------------------------------------------------------
Input:
Grid with A (snake A start), B (snake B start), . (free), # (wall)

Output:
Step-by-step movement log and final winner based on survival time.

Execution:
Give grid size + grid rows. Snakes move automatically.
-----------------------------------------------------------------
"""

import random

# ------------- USER INPUT -------------
r, c = map(int, input("Enter rows and columns: ").split())
print("Enter grid row by row:")
grid = [list(input()) for _ in range(r)]

for i in range(r):
    for j in range(c):
        if grid[i][j] == 'A': A = (i, j)
        if grid[i][j] == 'B': B = (i, j)

DIR = [(0,1),(1,0),(0,-1),(-1,0)]   # Right, Down, Left, Up
dA = 0; dB = 2                      # face each other
bodyA = {A}; bodyB = {B}
scoreA = scoreB = 0
log = []


def move(pos, d): return (pos[0]+DIR[d][0], pos[1]+DIR[d][1])


def safe(npos, my, other):
    return 0 <= npos[0] < r and 0 <= npos[1] < c and grid[npos[0]][npos[1]] != '#' and npos not in my and npos not in other


def choose(pos, d, my, other):
    f = d
    l = (d - 1) % 4
    rt = (d + 1) % 4
    for nd in [f, l, rt, random.randint(0,3)]:
        np = move(pos, nd)
        if safe(np, my, other): return nd
    return d


# ---------------- GAME ----------------
winner = None
for t in range(1, 200):
    dA = choose(A, dA, bodyA, bodyB)
    dB = choose(B, dB, bodyB, bodyA)

    A = move(A, dA)
    B = move(B, dB)

    deadA = not safe(A, bodyA, bodyB)
    deadB = not safe(B, bodyB, bodyA)

    if not deadA: scoreA += 1; bodyA.add(A)
    if not deadB: scoreB += 1; bodyB.add(B)

    log.append(f"Step {t}: A ‚Üí {A} | B ‚Üí {B}")

    if deadA and deadB: winner = "Draw"; break
    if deadA: winner = "Agent B wins"; break
    if deadB: winner = "Agent A wins"; break

if winner is None:
    winner = "Agent A wins" if scoreA > scoreB else "Agent B wins" if scoreB > scoreA else "Draw"

# --------------- OUTPUT ---------------
print("\n------ SMART SNAKE AI DUEL LOG ------")
for entry in log[:30]: print(entry)
if len(log) > 30: print("... (more steps hidden) ...")

print("\nScores:")
print("Agent A survival score:", scoreA)
print("Agent B survival score:", scoreB)
print("üèÜ RESULT:", winner)
```
